```python
# Step 1: Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Step 2: Create dataset (hospital data)
data = {
    'Age': [25, 45, 35, 50, 23, 40, 60, 48],
    'BP': [120, 140, 130, 150, 110, 135, 160, 145],
    'Cholesterol': [200, 240, 220, 260, 180, 230, 280, 250],
    'Sugar': [85, 90, 88, 95, 80, 92, 100, 96],
    'Disease': [0, 1, 0, 1, 0, 1, 1, 1]
}

df = pd.DataFrame(data)

# Step 3: Split data
X = df[['Age', 'BP', 'Cholesterol', 'Sugar']]
y = df['Disease']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Step 4: Train model
model = LogisticRegression()
model.fit(X_train, y_train)

# Step 5: Predict
y_pred = model.predict(X_test)

# Step 6: Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Step 7: Test new patient
new_patient = pd.DataFrame([[30, 130, 210, 88]], columns=['Age', 'BP', 'Cholesterol', 'Sugar'])
prediction = model.predict(new_patient)

if prediction[0] == 1:
    print("Patient has disease")
else:
    print("Patient is healthy")
```
