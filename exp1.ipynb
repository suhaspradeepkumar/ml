```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("exp1.csv")

# Show first few rows
df.head()
```

```python
def compute_stats(series):
    return {
        "Mean": series.mean(),
        "Median": series.median(),
        "Mode": series.mode().tolist(),
        "Std Dev": series.std(),
        "Variance": series.var()
    }

temp_stats = compute_stats(df["Temperature"])
hum_stats  = compute_stats(df["Humidity"])

print("Temperature Stats:\n", temp_stats)
print("\nHumidity Stats:\n", hum_stats)

```

```python
plt.figure()
plt.scatter(df["Temperature"], df["Humidity"])
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Scatter Plot: Temperature vs Humidity")
plt.show()
```

```python
plt.figure()
plt.plot(df["Temperature"], label="Temperature")
plt.plot(df["Humidity"], label="Humidity")
plt.legend()
plt.title("Original Data (Before Normalization)")
plt.xlabel("Sample Index")
plt.ylabel("Value")
plt.show()
```

```python
# MinMax Normalization
df["Temp_MinMax"] = (df["Temperature"] - df["Temperature"].min()) / (df["Temperature"].max() - df["Temperature"].min())
df["Hum_MinMax"]  = (df["Humidity"] - df["Humidity"].min()) / (df["Humidity"].max() - df["Humidity"].min())
```

```python
plt.figure()
plt.plot(df["Temp_MinMax"], label="Temp Min-Max")
plt.plot(df["Hum_MinMax"], label="Hum Min-Max")
plt.legend()
plt.title("MinMax Normalized Data")
plt.xlabel("Sample Index")
plt.ylabel("Normalized Value")
plt.show()
```

```python
plt.figure()
plt.plot(df["Temperature"], label="Original")
plt.plot(df["Temp_Z"], label="Standardized (Z)")
plt.legend()
plt.title("Temperature: Original vs Standardized")
plt.xlabel("Sample Index")
plt.ylabel("Value / Z-score")
plt.show()
```

```python
plt.figure()
plt.plot(df["Humidity"], label="Original")
plt.plot(df["Hum_Z"], label="Standardized (Z)")
plt.legend()
plt.title("Humidity: Original vs Standardized")
plt.xlabel("Sample Index")
plt.ylabel("Value / Z-score")
plt.show()
```
